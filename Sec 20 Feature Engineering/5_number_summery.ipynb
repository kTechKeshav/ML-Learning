## 5 number Summary and Box Plot

```python
## Minimum, Maximum, Q1, Q3, IQR
```

```python
import numpy as np
```

```python
import random
lst_marks = [37, 18, 48, 29, 49, 35, 94, 26, 48, 29, 48, 37, 29]

min, Q1, median, Q3, max = np.quantile(lst_marks, [0, 0.25, 0.50, 0.75, 1.0])

print(min)
print(Q1)
print(median)
print(Q3)
print(max)
```

```python
IQR = Q3-Q1
print(IQR)
```

```python
lower_fence = Q1- (1.5*IQR)
print("Lower Fence ",lower_fence)
higher_fence = Q3 + (1.5*IQR)
print("Upper Fence ",higher_fence)

outlier = [x for x in lst_marks if x<lower_fence or x>higher_fence]
outlier
```

```python
import seaborn as sns

sns.boxplot(lst_marks)
```
