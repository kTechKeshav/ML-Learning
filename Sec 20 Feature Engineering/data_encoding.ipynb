#### Nominal/ OHE Encoding

1. Red [1, 0, 1]
2. Green [0, 1, 0]
3. Blue [0, 0, 1]

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.DataFrame({
      'color': ['pink', 'red', 'blue', 'green', 'green', 'red', 'blue']
})

df
```

```python
## Create an instance of OneHotEncoder

encoder = OneHotEncoder()
```

```python
## Perform fit and than transform
encoded=encoder.fit_transform(df[['color']]).toarray()
encoded
```

```python
import pandas as pd
encoder_df=pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
encoder_df
```

```python
pd.concat([df, encoder_df], axis=1)
```

#### Label Encoding

1. Red 1
2. Green 2
3. Blue 3

```python
import pandas as pd

df = pd.DataFrame({
      'color': ['pink', 'red', 'blue', 'green', 'green', 'red', 'blue']
})

df
```

```python
from sklearn.preprocessing import LabelEncoder

lbl_encoder = LabelEncoder()
```

```python
encoded = lbl_encoder.fit_transform(df[['color']])
encoded
```

### Ordinal Encoding

1. High school: 1
2. College: 2
3. Graduate: 3
4. Post-Graduate: 4

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

df = pd.DataFrame({
      'size': ['small', 'medium', 'large', 'medium', 'small', 'large']
})

df
```

```python
ord_encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
```

```python
encoded = ord_encoder.fit_transform(df[['size']])
encoded
```

### Target Guided Ordinal Encoding

```python
import pandas as pd

df = pd.DataFrame({
      'city': ['New York', 'London', 'Paris', 'Tokyo', 'New York', 'Paris'],
      'price': [200, 150, 300, 250, 180, 320]
})

df
```

```python
mean_price=df.groupby('city')['price'].mean().to_dict()
```

```python
mean_price
```

```python
df['city_encoded']=df['city'].map(mean_price)
# df[['city', 'city_encoded']]
df
```
